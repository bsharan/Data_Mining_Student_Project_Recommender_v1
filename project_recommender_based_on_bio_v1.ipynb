```python
# now not frequent but important words

text = [bio]

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')

bio_tfidf = vectorizer.fit_transform(text)

```

```python
import pandas as pd

df = pd.DataFrame(bio_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
# Show top words
df = df.T.sort_values(by=0, ascending=False)
print("Top keywords from my bio:")
print(df.head(10))
```
